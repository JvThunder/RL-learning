```python
import gym
import gym_classics
import warnings
import numpy as np
import random
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
env = gym.make("WindyGridworld-v0")
params = {
    "alpha": 0.5,
    "gamma": 0.99,
    "epsilon": 0.1,
}
EPISODES = 500
Q = np.random.uniform(low=0, high=1, size=(62, 4))
reward_list = []
for i in range(1, EPISODES + 1):
    done = False
    state = env.reset()
    action = env.action_space.sample()
    tot_reward = 0
    while not done:
        next_state, reward, done, info = env.step(action)
        
        if random.uniform(0,1) < params["epsilon"]:
            next_action = env.action_space.sample()
        else:
            next_action = np.argmax(Q[state,:])

        td_error = 0
        if not done:
            td_error = reward + params["gamma"] * Q[next_state, next_action] -  Q[state, action]
        else:
            td_error = reward -  Q[state, action]
        Q[state, action] += params["alpha"] * td_error

        action = next_action
        state = next_state
        tot_reward += reward
    
    reward_list.append(tot_reward)
```

```python
plt.plot(reward_list)
plt.show()
```
